```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s/ref=nb_sb_noss_1?url=search-alias%3Daps&field-keywords=headphone+sony&crid=1M0KDR7SG9MGE&sprefix=headphone+sony,aps,318"
```

```python
# Headers for request
HEADERS = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://amazon.com" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containiang all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```
